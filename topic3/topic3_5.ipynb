#### 미니프로젝트 10. 자료구조 queue 구현

- 일종의 대기줄

#### pseudo code

- 리스트 만듦
- 명령어 입력
    - push
        - 숫자 입력
        - 리스트 push
    - out
        - remove(arr[0])
        - pop(0)
    
- 리스트 출력

#### queue

- 삽입과 삭제가 각각 다른 편에서 이루어지는 자료구조
- 일종의 대기줄
- FIFO (First in, First out)

```python
queue = []

while True : 
    cmd = input("명령어를 입력하세요 : ")
    
    match (cmd) :
        case 'push' : 
            num = int(input("숫자를 입력하세요 : "))
            queue.append (num)
            
        case 'out' : 
            queue.pop(0)
            
        case 'exit' : 
            break
    
    print (queue)
```
