##### 크기가 다름을 방지 == squared matrix

**pseudo code**
- squared matrix의 행의 갯수 N
- N만큼 반복하여 row씩 component들을 받음
- 사용자가 +, -, *, c를 입력하면
    - \+ : 행렬의 덧셈
    - \- : 행렬의 뺄셈
    - \* : 행렬의 곱셈
    - c : 상수를 입력받음 -> 행렬의 상수배

```python
N = int(input())

A = []
B = []

for i in range (N) : # A 입력 (row 단위)
    row = list(map(int, input().split())) # 입력받은 것을 split한다 (공백을 기준으로) -> map함수로 int를 각각 지정한다
    A.append (row)
    
for i in range (N) : # B 입력 (row 단위)
    row = list(map(int, input().split())) # 입력받은 것을 split한다 (공백을 기준으로) -> map함수로 int를 각각 지정한다
    B.append (row)
    
print (A)
print (B)
```

```python
operator = input('operator를 입력하세요 : ')

'''if operator == '+' : 
    # 행렬의 덧셈
    
elif operator == '-' :
    # 행렬의 뺄셈
    
elif operator == '*' :
    # 행렬의 곱셈
    
elif operator == 'c' :
    # 행렬의 상수배
    
else : 
    # 오류 출력 및 프로그램 종료'''

C = []
    
match (operator) :
    case '+' :
        for i in range (N) : # row
            row = []
            
            for j in range (N) : # col
                row.append (A[i][j] + B[i][j])
                
            C.append(row)

    case '-' :
        for i in range (N) : # row
            row = []
            
            for j in range (N) : # col
                row.append (A[i][j] - B[i][j])
                
            C.append(row)
        
    case '*' :
        '''이 부분을 구현하는 것이 과제입니다'''
        pass
        
    case 'c' : 
        obj = input("어떤 행렬을 상수배하실 겁니까? : ")
        constant = int(input("상수는 무엇입니까? : "))
        
        if obj == 'A' : 
            for i in range (N) : # row
                row = []

                for j in range (N) : # col
                    row.append (A[i][j] * constant)

                C.append(row)
        
        elif obj == 'B' : 
            for i in range (N) : # row
                row = []

                for j in range (N) : # col
                    row.append (B[i][j] * constant)

                C.append(row)

print (C)
```
