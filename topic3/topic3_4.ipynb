##### pseudo code
- 10진수를 2진수로 변환할 수 N를 입력
- 몫 = -1
- 나머지 = []
- 반복문 (몫이 0이 아닐 때까지)
    - 나머지.append(N % 2)
    - N = N // 2 
- 나머지.append (N % 2)

```python
import sys 

N = int(input("2진수로 변환할 수 N을 입력 : ")) # 2진수로 변환할 수 N을 입력
div = -1
rest = []

while N != 0 : 
    rest.append (N % 2)
    N = N // 2
    
rest.append (N % 2)
rest.pop()

binary = rest[::-1]

for i in range (len(binary)) :
    binary[i] = str(binary[i])

binary_word = ''.join(binary)
print (binary_word)
```

```python
# 5! : 5 * 4 * 3 * 2 * 1
# 10! : 10 * 9 * 8 * ... * 3 * 2 * 1
# n! : n * (n-1) * (n-2) ... * 3 * 2 * 1

n = int(input("n 펙토리얼에서 n을 입력 : "))

rst = 1

while n != 0 : 
    rst *= n
    n -= 1
    
print (rst)
```

```python
# 재귀 : 자신의 함수 안에 자신의 함수를 한번 더 사용하는 것 (함수를 연속적으로 call stack에 저장해가면서 자신을 계속 호출하는 것)

def factorial (n) :
    if n == 0 : 
        return 1
    
    else : 
        return n * factorial(n-1)

n = int(input("n 펙토리얼에서 n을 입력 : "))
rst = factorial(n)

print (rst)
```
