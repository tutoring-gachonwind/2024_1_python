#### 자료구조 (2학년 1학기 / C언어)
- 데이터를 효과적으로 분류/처리/수납/삭제 할 수 있는 일종의 공간
    - 선형구조 : stack, queue, deque (리스트)
    - 비선형구조 : tree, graph (특정 기준에 맞춰서 저장 / set)

#### stack
- 

- stack == 돌탑
    - 처음에 돌을 쌓는다 - 아래서부터 차례대로 쌓는다 (append / push)
    - 돌을 뺀다 - 마지막으로 올렸던 돌부터 빼기 (pop)
    
- FIFO (First in, First out) : push한 방향 == 반대(pop한 방향)

--------

#### pseudo code

- 리스트 만듦
- 명령어 입력
    - push
        - 숫자 입력
        - 리스트 push
    - pop
        - pop
    
- 리스트 출력

```python
stack = []

while True :
    command = input("명령어를 입력하세요 : ")
    
    if command == "push" :
        num = int(input("숫자를 입력하세요 : "))
        stack.append (num)
        
    elif command == "pop" :
        stack.pop()
        
    elif command == "exit" :
        break
        
    print ("stack의 현 상태 : ", stack)
```
